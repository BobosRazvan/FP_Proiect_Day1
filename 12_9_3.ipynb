```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['pricerunner_aggregate.csv']))
```

```python
df.head()
```

```python
category_labels = df[' Category Label'].unique()
```

```python
category_labels
```

```python
df_grouped = df.groupby(' Category Label')
```

```python
sampled_df = df_grouped.head(15).reset_index(drop=True)
```

```python
sampled_df.shape
```

```python
sampled_df.head(50)
```

```python
X_data = sampled_df[['Product Title', ' Cluster Label']]
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

encoded_data = encoder.fit_transform(X_data[['Product Title', ' Cluster Label']])

encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['Product Title', ' Cluster Label']))
```

```python
X_data = pd.concat([X_data, encoded_df], axis=1)

X_data = X_data.drop(columns=['Product Title', ' Cluster Label'])
```

```python
X_data.head()
```

```python
df_united = pd.concat([sampled_df, X_data], axis=1)
```

```python
df_united.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df_united[[' Merchant ID', ' Cluster ID', ' Category ID']])
df_united[[' Merchant ID', ' Cluster ID', ' Category ID']] = scaler.transform(df_united[[' Merchant ID', ' Cluster ID', ' Category ID']])
df_united.head()
```

```python
df = df_united.drop(columns=[ 'Product ID', 'Product Title', ' Cluster Label', ' Category Label'])
df.head()
```

```python
df.to_csv('input_razvan.csv', header=False, index=False)
```
